```python
# Load CRAM paths on ICA
fn = '/DC3/AGD250k_CRAM_info_only/agd250k_file_path_on_ica.txt'
df_path = pd.read_csv(fn, sep='\t').rename(columns={'file_path':'cram_path'})
print(df_path.shape)
df_path['crai_path'] = df_path['cram_path']+'.crai'
display(df_path.head(2))
```

```python
output_fh = open('./bash_slurm/extract_hla_reads.cmd_run_ica.extract_hla_reads_workspace.sh', 'w')
n_samples_per_output_folder = 5000
# This should be chaning based on the number of samples
# output_path_prefix = '/data/project/wanying/kourami_hla_typing/chr6_hla_bams/workspace_extraction_'
output_path_prefix = '/data/wanying/data/workspace_extraction_'
c = 0
# /data/wanying/code/utils/extract_hla_reads_from_cram.sh
for cmd in 'bash /data/wanying/code/utils/extract_hla_reads_from_cram.no_sorting.sh "' + df_path['sample_id'] + '" "/data/project' + df_path['cram_path'] + '" "/data/project' + df_path['crai_path']+ '" ':
    folder_indx = c//n_samples_per_output_folder + 1
    output_path = f'{output_path_prefix}{(folder_indx-1)*n_samples_per_output_folder+1}-{folder_indx*n_samples_per_output_folder}'
    log_file = f'/data/wanying/data/workspace_extraction_{(folder_indx-1)*n_samples_per_output_folder+1}-{folder_indx*n_samples_per_output_folder}.log'
    output_fh.write(cmd + output_path + ' ' + log_file + '\n')
    c += 1
    if c%100==0:
        print(f'\r# Total cmd written: {c}', flush=True, end='')
print(f'\r# Total cmd written: {c}')
output_fh.close()
```
