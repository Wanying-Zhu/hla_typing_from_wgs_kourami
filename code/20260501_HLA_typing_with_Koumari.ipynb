```python
%%bash
# Do it in terminal
# Start the Docker container and run from it.
# Autumn's NF pipeline example: /data100t1/home/armorrow/als_ftd/genetics/McMillan_collab/code/nextflow_combineVCFs.nf

docker run --name kourami_hla \
--mount type=bind,src=/data100t1/home/wanying/BioVU/20260303_AGD_250k_HLA_imputation/data/CRAM_example,dst=/data/crams,readonly \
--mount type=bind,src=/data100t1/home/wanying/shared_data_files/Illumina_reference,dst=/data/reference,readonly \
--mount type=bind,src=/data100t1/home/wanying/BioVU/20260303_AGD_250k_HLA_imputation/output,dst=/data/output \
--mount type=bind,src=/data100t1/home/wanying/BioVU/20260303_AGD_250k_HLA_imputation/code,dst=/data/code \
-it kourami-hla-tools:1.0 \
/bin/sh

# Now the files are stored at:
# - Example CRAM: /data/crams/R299996754.cram
# - Reference for CRAM: /data/reference/GRCh38_full_analysis_set_plus_decoy_hla.fa
# - Output files: /data/output

# Tools are stored in /usr/local/bin/
# - Kourami: /usr/local/kourami-0.9.6/build/Kourami.jar
# - gatk: /usr/local/gatk/gatk (or gatk.jar)
```
