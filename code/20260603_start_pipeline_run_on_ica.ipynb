```python
fn_file_ids = '/DC3/AGD250k_CRAM_info_only/agd_cram_file_ids_from_2025_q1_release.na_fixed.tsv'
n_samples_per_batch = 10

# Output to /wanying/kourami_hla_typing/chr6_hla_bams/
# Based on the test run, 10 samples is a good number and reliable
n_samples_per_batch = 10

cmd = 'pipeline_id=156855c9-d3c3-4a64-9768-5a45d0b7f027; output_parent_folder_id=fol.712f6fae5f3c416e66b408debcedb98c; cram_ref_file_id=fil.d0a8266ac310465a5fff08deaaf769c0; /data100t1/home/wanying/downloaded_tools/icav2/icav2 projectpipelines start nextflow ${pipeline_id} '
count = 0
# Save logs to extraxt id of each analysis, and then id of the output bams (for downloading)
log_output_path = '/DC3/AGD250k_HLA_typing/extracted_hla_read_bams/ica_submission_logs'
output_fn = './bash_slurm/icav2_pipeline_run.extract_hla_reads.sh'
output_fh = open(output_fn, 'w')
with pd.read_csv(fn_file_ids, sep='\t', chunksize=n_samples_per_batch) as reader:
    for chunk in reader:
        cram_ids = '"'+ '","'.join(chunk['cram_id']) + '"'
        cram_index_ids = '"'+ '","'.join(chunk['cram_index_id']) + '"'
        cmd_output = cmd + '--input cram_path:' + cram_ids + ' '
        cmd_output += '--input cram_index:' + cram_index_ids + ' '
        cmd_output += '--input cram_ref:${cram_ref_file_id} '
        cmd_output += '--output-parent-folder ${output_parent_folder_id} '
        cmd_output += '--storage-size small '
        cmd_output += f"--user-reference 'extract_hla_reads_run_{count+1}' > {log_output_path}/extract_hla_reads_run_{count+1}.ica_submission.log"
        output_fh.write(cmd_output + '\n')
        count += 1
        if count%1000==0:
            print(f'\r# Total commands written: {count}', end='', flush=True)
output_fh.close()
print(f'\r# Total commands written: {count}')
```
